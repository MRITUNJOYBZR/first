```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# read the dataset
df=pd.read_excel("HomePrice.xlsx")
df
```

```python
# extract the area column of the dataset
df.area
```

```python
# extract the price column of the dataset
df.price
```

```python
# relation bw area and price columns of the dataset before the prediction
%matplotlib inline
plt.scatter(df.area,df.price,color='red',marker="+")
plt.xlabel("area(sqr ft)")
plt.ylabel('price(US $)')
plt.grid(True)
plt.show()
```

```python
# initialize the linear regression model
reg=linear_model.LinearRegression()
```

```python
# make the numpy array of the area column of the dataset
df[['area']].values
```

```python
# make the numpy array of the price column of the dataset
df[['price']].values
```

```python
# fit the model
reg.fit(df[['area']].values,df[['price']].values)
```

```python
# prediction
reg.predict([[3300]])
```

```python
# coefficient
reg.coef_
```

```python
# intercept
reg.intercept_
```
